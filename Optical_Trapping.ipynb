```python
laser=[100,125,150,175,200]
for lz in laser:
    print(f"Laser: {lz}")
    data = pd.read_csv(f'x_{lz}_vl_bs.txt', sep='\t', 
                      names=["QPD X", "QPD Y", "Piezo X", "Piezo Y"])
    fig,ax = plt.subplots()
# make a plot
    ax.plot(timexs, data["Piezo Y"].values, label="Stage", color="red")
    # set x-axis label
    ax.set_xlabel("Time $(sec)$",fontsize=14)
    # set y-axis label
    ax.set_ylabel("Strain Gauge Y $(Volts)$",color="red",fontsize=14)
    # twin object for two different y-axis on the sample plot
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(timexs, data["QPD Y"].values, label="QPD Y")
    ax2.set_ylabel("QPD Y $(millivolts)$",color="blue",fontsize=14)
    plt.show()
```
