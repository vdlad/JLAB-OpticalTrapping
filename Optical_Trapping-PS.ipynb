```python
def power_spec(freq,fnot,A):
    return A/(freq**2+fnot**2)
```

# QPD X Data Analysis

```python
laser=[100,125,150,175,200]
for index in range(len(laser)):
    data = pd.read_csv(f'bm_vl_bs_{laser[index]}.txt', sep='\t', 
                  names=["QPD X", "QPD Y", "Piezo X", "Piezo Y"])
    
    f, Pxx_den = signal.welch(data["QPD X"].values*x_factor[index],5000,nperseg=50000-1)
    bin_means, bin_edges, binnumber = stats.binned_statistic(f,Pxx_den, statistic='mean', bins=25000/6)
    bin_std, bin_edges, binnumber=stats.binned_statistic(f,Pxx_den, statistic='std', bins=25000/6)
    
    plt.loglog(f, Pxx_den, label=str(laser[index]))
    plt.loglog(bin_edges[:-1],bin_means,'black', label="Binned")
#     plt.errorbar(bin_edges[:-1], bin_means,yerr=bin_std/np.sqrt(10), linewidth=1, ls='none', fmt='o', capsize=3,color="gray")
    
    popt, pcov = curve_fit(f=power_spec, xdata=bin_edges[:-1], ydata=bin_means, sigma=bin_std)
    
    
    r = bin_means - power_spec(bin_edges[:-1], *popt)
    chisq = sum((r / bin_std) ** 2)
    print(chisq)
    perr = np.sqrt(np.diag(pcov))
    print(perr)
    plt.plot(f, power_spec(f, popt[0], popt[1]), label="Curve Fitted")

    

    plt.xlabel('frequency [Hz]', fontsize=16)
    plt.ylabel('PSD $[\mu m^2/Hz]$', fontsize=16)
    plt.legend(fontsize=16)
    plt.show()
    
```

```python
def alpha(f_not):
    beta=3.0*np.pi*8.90*10**(-4)*3.71*10**(-6)
    return np.array([2*np.pi*beta])*f_not
```

```python
def boltzman(A):
    rho=1.99
    beta=3.0*np.pi*8.90*10**(-4)*3.71*10**(-6)
    num=np.pi**2*beta
    denom=(rho**2)*293
    return [element*num/denom for element in A]
    
    
```
